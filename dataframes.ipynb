```python
df_before= df.copy()
```

```python
df['Units_Sold']=df['Units_Sold'].fillna(df['Units_Sold'].mean())
df['Discount_Percent']= df['Discount_Percent'].fillna(df['Discount_Percent'].mean())
df['Region']= df['Region'].replace(0,np.nan).fillna('South')
```

```python
print('Null count after cleaning: ')
print(df.isnull().sum())
```
